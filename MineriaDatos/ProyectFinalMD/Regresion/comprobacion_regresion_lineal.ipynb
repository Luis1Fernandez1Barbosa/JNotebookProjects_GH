## Comprobación de Regresión Lineal


```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

### Configuración de entorno para la importación de datos

```python
diabetes = datasets.load_diabetes()
#insert una nueva dimensión y selecciona la tercera columna
y = diabetes.target
X = diabetes.data[:, np.newaxis, 2]
print(diabetes.data.shape)
print(diabetes.data[:2].shape)
print(diabetes.data[:,np.newaxis,2].shape)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10)
```

### Regresión Lineal: entretamiento y predicción

```python

#obtiene una instancia de la clase LinearRegression
linear_regression = linear_model.LinearRegression()
#crea el modelo utilizando los datos de entrenamiento
linear_regression.fit(X_train, y_train)

y_train_predicted = linear_regression.predict(X_train) # predice los valores de "y" para los datos de entrenamiento
y_test_predicted = linear_regression.predict(X_test) # predice los valores de "y" para los datos de test

#conjunto de datos obtenido con el entrenamiento
train_MSD = mean_squared_error(y_train, y_train_predicted)
#conjunto de datos obtenido con el test
test_MSD = mean_squared_error(y_test, y_test_predicted)
```

### Visualización de los resultados

```python
#dibujando el entrenamiento y el test
fig, axs = plt.subplots(1,2, figsize=(15,4))
#axs[0] es el primer gráfico con los datos de entrenamiento
axs[0].scatter(X_train, y_train, color='orange')
axs[0].plot(X_train, y_train_predicted, color='black')
axs[0].set_title('Entrenamiento conjunto, MSD:{:0f}'.format(train_MSD))
#axs[1] es el segundo gráfico con los datos de test
axs[1].scatter(X_test, y_test, color='gray')
axs[1].plot(X_test, y_test_predicted, color='black')
axs[1].set_title('Test conjunto, MSD:{:0f}'.format(test_MSD))
```
